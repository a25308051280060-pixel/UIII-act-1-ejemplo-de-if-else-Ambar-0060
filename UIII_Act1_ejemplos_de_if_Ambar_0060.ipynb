#### ejemplo de if

```python
# A. Ejemplo de if simple (Relacional y Aritmético)
# Este ejemplo verifica si un número es par utilizando el operador de módulo.
numero = 10
if numero % 2 == 0:
    print("El número es par.")
```

```python
# B. Ejemplo de if else (Relacional y Lógico)
# Este ejemplo determina el acceso a un lugar basándose en la edad y una autorización.
edad = 20
autorizado = True
if edad >= 18 and autorizado:
    print("Acceso concedido.")
else:
    print("Acceso denegado.")
```

```python
# C. Ejemplo de if elif (Aritmético, Relacional y Lógico)
# Este ejemplo clasifica el rendimiento académico de un estudiante calculando un promedio.
puntos = 85
asistencia = 90
if puntos >= 90:
    print("Rendimiento: Sobresaliente")
elif puntos >= 80 and asistencia >= 85:
    print("Rendimiento: Muy bueno")
elif puntos >= 70 or asistencia >= 80:
    print("Rendimiento: Aceptable")
else:
    print("Rendimiento: Insuficiente")
```

```python
# D. Ejemplo Complejo: Año Bisiesto (Aritmético, Relacional y Lógico)
# Verifica si un año ingresado por el usuario es bisiesto siguiendo reglas matemáticas específicas
año = 2024
if (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0):
    print(f"{año} es un año bisiesto.")
else:
    print(f"{año} no es un año bisiesto.")
```
